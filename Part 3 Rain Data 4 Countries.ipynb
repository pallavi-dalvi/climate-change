```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# This magic line will allow you to generate plots
# within the Jupyter notebook.
%matplotlib inline

#this adds some extra styling to the graphs we will make
sns.set_style("whitegrid")
```

```python
df1=pd.read_csv("Rainfall_1901_2016_CHN.csv")
df2=pd.read_csv("Rainfall_1901_2016_IND.csv")
df3=pd.read_csv("Rainfall_1901_2016_BTN.csv")
df4=pd.read_csv("Rainfall_1901_2016_MDV.csv")

```

```python
df1.columns=['Rainfall - (MM)', 'Year', 'Month', 'Country', ' ISO3']
df1
```

```python
average_yearly_rain_china=df1[['Country','Rainfall - (MM)', 'Year']].groupby('Year',as_index=False).mean()
```

```python
average_yearly_rain_china['Country']='China'
```

```python
average_yearly_rain_china
```

```python
df2.columns=['Rainfall - (MM)', 'Year', 'Month', 'Country', ' ISO3']
df3.columns=['Rainfall - (MM)', 'Year', 'Month', 'Country', ' ISO3']
df4.columns=['Rainfall - (MM)', 'Year', 'Month', 'Country', ' ISO3']




average_yearly_rain_india=df2[['Country','Rainfall - (MM)', 'Year']].groupby('Year',as_index=False).mean()
average_yearly_rain_bhutan=df3[['Country','Rainfall - (MM)', 'Year']].groupby('Year',as_index=False).mean()
average_yearly_rain_maldi=df4[['Country','Rainfall - (MM)', 'Year']].groupby('Year',as_index=False).mean()
```

```python
average_yearly_rain_india['Country']='India'
average_yearly_rain_bhutan['Country']='Bhutan'
average_yearly_rain_maldi['Country']='maldi'
```

```python
average_yearly_rain_china.columns=['Year', 'Avg Rainfall - (MM)', 'Country']
average_yearly_rain_india.columns=['Year', 'Avg Rainfall - (MM)', 'Country']
average_yearly_rain_bhutan.columns=['Year', 'Avg Rainfall - (MM)', 'Country']
average_yearly_rain_maldi.columns=['Year', 'Avg Rainfall - (MM)', 'Country']
```

```python
average_yearly_rain_india
```

```python
x= pd.concat([average_yearly_rain_china,average_yearly_rain_india,average_yearly_rain_bhutan,average_yearly_rain_maldi])
```
